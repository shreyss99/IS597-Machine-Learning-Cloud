# Exercise 1

With the data provided below, create one linear plot.    
- Data: two lists containing group numbers and study hours for each group.
- The output plot must meet the following requirements:  
  A. The plot must contain a title named "Study Hours Per Group".  
  B. The plot must contain a legend with label named 'Linear Plot', which is located at the upper left corner in the figure.   
  C. The plot must contain a line style as follows: color (purple), line width (2), marker ('+'), marker size (10).   
  D. The plot must contain X and Y label that correspond the name of lists (groups, hours) where data is included.  
  E. The plot must contain the X limit and y limit that have the range of (0,8) and (0,15), respectively.   
  F. The plot must be displayed in the Jupyter Notebook itself as well as saved in a separate output file named 'exercise1.png'.  



```python
# RUN THIS CELL - Data provided

groups = [1, 2, 3, 4, 5, 6, 7]
hours  = [2, 4, 6, 8, 10, 12, 14]
```

```python
# Your code here

# Plot
plt.plot(groups, hours, color='purple', linewidth=2, marker='+', markersize=10, label='Linear Plot')
plt.title('Study Hours Per Group')
plt.xlabel('Groups')
plt.ylabel('Hours')
plt.legend(loc='upper left')
plt.xlim(0, 8)
plt.ylim(0, 15)
plt.grid(True)

# Display and save the plot
plt.savefig('exercise1.png')
plt.show()
```

# Exercise 2 (Challenge)

Create a figure that contains 2 subplots located side by side.    
- Data: two sets of data are provided for use.
- The output figure resulted from your code should meet the following requirements:  
A. Figure size is set to (5,4) with a dpi of 200.  
B. The figure must contain a title named 'Two Plots: A & B".  
C. Each subplot must have a title as follows: 'Plot A' & 'Plot B'.  
D. Each subplot must display X and Y labels that correspond to the name of data used.  
E. Each subplot must contain a legend at the location of lower right corner in the plot with label named 'Linear' and 'Exp', respectively.   
F. Each plot must have its own line style and color as follows: color (red & lightblue), shape (dash & dash-dot), width (1 for both plots), marker ('o' & '+'), marker size (2 & 4).  
G. Both plots must contain the same range of x limit and y limit as follows: x axis (0,11) & y axis (0,105).  
H. Make sure the two plots do not overlap each other due to the lack of space between them.  
I. In addition to displaying the figure in the Jupyter notebook, save it into a separate file with the name 'exercise2.png'. Make sure that any parts of this figure are not cut off when you open the file (you can do this by setting a parameter in the code used for saving).



```python
# RUN THIS CELL - Starting point

# Data for Plot A
X1 = np.linspace(0,10,11)
Y1 = X1 * 10

# Data for Plot B
X2 = np.linspace(0,10,11)
Y2 = X2 ** 2
```

```python
# Your code here

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(5, 4), dpi=200)

# Plot A
axs[0].plot(X1, Y1, color='red', linestyle='--', linewidth=1, marker='o', markersize=2, label='Linear')
axs[0].set_title('Plot A')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].legend(loc='lower right')
axs[0].set_xlim(0, 11)
axs[0].set_ylim(0, 105)

# Plot B
axs[1].plot(X2, Y2, color='lightblue', linestyle='-.', linewidth=1, marker='+', markersize=4, label='Exp')
axs[1].set_title('Plot B')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].legend(loc='lower right')
axs[1].set_xlim(0, 11)
axs[1].set_ylim(0, 105)

# Figure Title
plt.suptitle('Two Plots: A & B')

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the figure
plt.savefig('exercise2.png', bbox_inches='tight')

# Display the figure
plt.show()
```
